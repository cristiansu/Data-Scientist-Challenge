```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
models = {
    'Logistic Regression' : LogisticRegression(),
    'KNN' : KNeighborsClassifier(n_neighbors=10),
    'Random': RandomForestClassifier(n_estimators=400)
}
```

```python
def fit_and_score(models, X_train, X_test, y_train, y_test):
    """
    Función que se encarga de entrenar y 
    obtener el puntaje de cada modelo entrenado
    """
    np.random.seed(42)
    #guardar los scores 
    models_scores = {}
    
    #loop
    for name, model in models.items():
        #entrenamos el modelo
        model.fit(X_train, y_train)
        #evaluamos el desempeño
        models_scores[name] = model.score(x_test, y_test)
        
    return models_scores
```

```python
models_scores = fit_and_score(models, x_train, x_test, y_train, y_test)
```
