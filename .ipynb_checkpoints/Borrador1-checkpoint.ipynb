```python
def temporada_alta(mes,dia):
    for m,d in zip(mes,dia):
        if m==1 or m==2:
            res=1
        elif m==3 & (1<=d<=3):
            res=1
        elif (m==7) & (15<=d<=31):
            res=1
        elif (m==9) & (11<=d<=30):
            res=1
        elif (m==12) & (d>=15):
            res=1
        else:
            res=0
    return res
```

```python
df['temporada_alta']=temporada_alta(df['MES'],df['DIA'])
```

```python
import datetime
from datetime import timedelta
def atraso_min(df):
    limite=df['Fecha-I']+datetime.timedelta(minutes=15)
    if df['Fecha-O']>limite:
        res=1
    else:
        res=0
    return res
    
```

```python
#------------NO FUNCIONA------------------
import time 
def diferencia(df):
    fmt = '%Y-%m-%d %H:%M:%S' 
    f1 = datetime.datetime.strptime(str(df['Fecha-O']), fmt) 
    f2 = datetime.datetime.strptime(str(df['Fecha-I']), fmt) 
    
    f1_ts = time.mktime(f1.timetuple()) 
    f2_ts = time.mktime(f2.timetuple()) 
    res_min=(int(f1_ts-f2_ts)) / 60
   
    return res_min
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
models = {
    'Logistic Regression' : LogisticRegression(max_iter=1000),
    'KNN' : KNeighborsClassifier(n_neighbors=10),
    'Random': RandomForestClassifier(n_estimators=10)
}
```

```python
# creamos función para el entrenamiento 
def fit_and_score(models, X_train, X_test, y_train, y_test):
    """
    Función que se encarga de entrenar y 
    obtener el puntaje de cada modelo entrenado
    """
    np.random.seed(42)
    #guardar los scores 
    models_scores = {}
    
    #loop
    for name, model in models.items():
        #entrenamos el modelo
        model.fit(X_train, y_train)
        #evaluamos el desempeño
        models_scores[name] = model.score(x_test, y_test)
        
    return models_scores
```

```python
models_scores = fit_and_score(models, x_train, x_test, y_train, y_test)
```
